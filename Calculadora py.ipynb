```python
num1 = int(input ('coloque o primeiro numero'))
num2 = int(input ('coloque o segundo numero'))

op = int(input('Escolha a operação (1-soma, 2-subtração, 3-multiplicação, 4-divisão): '))

if op == 1:
    resultado = num1 + num2
    print (resultado)
elif op == 2:
    resultado = num1 - num2
    print (resultado)
elif op == 3:
    resultado = num1 * num2
    print (resultado)
elif op == 4:
    if num2 == 0:
     resultado = num1 / num2
     print (resultado)
else:
    print ('opção invalida')
```

# Estrutura de repetição

```python
# For range
for contador in range(10):
    print(contador)
```

```python
# For range + break
for contador in range(1000):
    if contador == 35:
        break
    print(contador)
```

```python
lista = ['teste1', 'teste2']
```

```python
# For simples
for val in lista:
    print(val)
```

```python
# For com enumerate
for i,val in enumerate(lista):
  print(i,val)
```

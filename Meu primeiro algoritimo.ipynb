## Tabela verdade

And

```python
print(10 > 2 and 4 < 6) # v-v = v
print(10 > 2 and 4 > 6) # v-f = f
print(10 < 2 and 4 < 6) # v-f = f
print(10 < 2 and 4 > 6) # f-f = f
```

Or

```python
print(10 > 2 or 4 < 6) #  v-v =v
print(10 > 2 or 4 > 6) #  v-f =v
print(10 < 2 or 4 < 6) #  f-v =v
print(10 < 2 or 4 > 6) #  f-f= f
```

not

```python
not 10 < 2
```

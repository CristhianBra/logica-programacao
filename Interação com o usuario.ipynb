# condicional

```python
#if
if 5 < 10:
    print('verdadeiro')
```

```python
#if/else

num1 = 10
num2 = 5
if num1 > num2:
    print('o num1 é o maior numero')
else:
    print('o num2 é o maior numero')
```

```python
#if/elif/else

num1 = 10
num2 = 10
if num1 > num2:
    print('o num1 é o maior numero')
elif num1 < num2:
    print('o num2 é o maior numero')
else:
    print('os numeros são iguais')
```

* condicional com operadores logicos

```python
num1 = 10
num2 = 5
num3 = 20
if num1 > num2 and num1 > num3:
    print('o num1 é o maior numero')
else:
    print('o num1 não é o maior numero')
```
